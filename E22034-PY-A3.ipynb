***1.WAP that will accept two sets of numbers and calculate the Pearson’s correlation 
coefficient between them. (Use necessary exception handling). (Optional if you do not 
know what it is)***

```python
num = int(input("Enter the size of the set: "))
x = []
y = []
for i in range(num):
    num_x = int(input("Enter the numbers for x: "))
    x.append(num_x)
for j in range(num):
    num_y = int(input("Enter the numbers for y: "))
    y.append(num_y)           


sum_x = 0
sum_y=0
z=0
rx=0
ry=0
prod=0

# Sum of elements in x 
for x_count in x:          
    sum_x = sum_x + x_count
# Sum of elements in y
for y_count in y:          
    sum_y = sum_y + y_count          
# Sum of product of x and y 
for n in range(0,len(x)):  
    prod=x[n]*y[n]
    z = z + prod
# Sum of squrt of elements in x    
for sqr_x in x:           
    x2=sqr_x**2
    rx = rx + x2
# Sum of squrt of elements in y    
for sqr_y in y:
    y2=sqr_y**2           
    ry = ry + y2

numerator = (len(x)*(z))-(sum_x*sum_y)
denominator = (((len(x)*rx)-(sum_x**2))*((len(x)*ry)-(sum_y**2)))**.5

result =numerator/denominator
print("Pearson’s correlation coefficient is ",result)

```

***2.Write a program to print all prime numbers from 1 to 300.***

***4.WAP which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.***

```python

count = 0
total = 0
    
while True:
    number = input("Enter a number: ")
    if number == 'done':
        break
    try:
        number1 = int(number)
    except:
        print("Invalid Input. ")
        continue
        
    total = total + number1
    count+=1
            
            

print("Total= ",total)
print("Count= ",count)
print("Average= ",total/count)
```

***5.Write another program that prompts for a list of numbers as above and at the end prints both the maximum and minimum of the numbers instead of the average.***

```python
sum1=0
count=0
max1=0
min1= " "
while True:
    num=input("Enter the number :")
    if num=="done":
        break
    try:
        num1=int(num)                             
        if (num1 > max1):                        
            max1=num1
        if (min1 is " " or num1 < min1):
            min1 = num1
    except:
        print("Invalid Input")
        continue
    sum1 = sum1 + num1
    count = count + 1

print(sum1,count,max1,min1)
```
